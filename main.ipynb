```python
import pandas as pd
import numpy as np
import model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from importlib import reload
```

```python
df=pd.read_csv('./Boston.csv')
df.head()
```

```python
X=df.drop(columns=["MEDV"]).to_numpy()
Y=df["MEDV"].to_numpy()
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=0.2)
```

```python
reload(model)
reg=model.regression(regularization="L2")
reg.Train(x_train,y_train,omega=10,calc_val=1,x_val=x_val,y_val=y_val,epoch=500,v=1)
```

```python
fig,ax=plt.subplots(figsize=(15,10))
ax.plot(reg.Loss_history["train"][1:],color="red",label="train")
ax.plot(reg.Loss_history["val"][1:],color="green",label="valid")
ax.set_xlabel("epochs")
ax.set_ylabel("loss")
ax.legend()
```
